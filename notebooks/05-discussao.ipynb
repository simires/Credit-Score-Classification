# Discussão dos Resultados

Este projeto teve como objetivo construir um modelo de Machine Learning para apoiar a classificação de pontuações de crédito de clientes a partir da base de dados [Credit score classification](https://www.kaggle.com/datasets/parisrohan/credit-score-classification/data). Ao longo do projeto, foram feitas as seguintes etapas: exploração inicial dos dados, limpeza e preparação, análise exploratória, pré-processamento, treinamento e comparação de três modelos de classificação (Regressão Logística, Random Forest e LightGBM), seleção e explicação do melhor modelo através do método SHAP. A seguir serão discutidos os resultados achados, recomendações práticas com base neles e respondidas as duas últimas perguntas de negócio norteadoras do projeto.

## Insights

A partir da análise exploratória dos dados, na procura de entender como as variáveis numéricas e categóricas estavam relacionadas com a classificação do score de crédito dos clientes, foi observado através dos boxplots e dos gráficos de barras agrupadas que o **comportamento financeiro** dos clientes era mais importante do que as características pessoais (como idade, profissão, entre outras). As variáveis que mais teriam poder discriminatório seriam taxa de juros, idade do histórico de créditos, dias de atraso de pagamento e dívida remanescente dos clientes. Ao treinar e testar o modelo e verificar a importância das variáveis, verificou-se que essas variáveis eram importantes, porém outras apareceram como mais importantes na classificação de Standard e Good. Assim, é observada a importância de não apenas interpretar gráficos na etapa de exploração, mas também de realizar investigações após o treinamento do modelo para entender o que realmente está sendo usado para a classificação.

Com relação aos modelos de Machine Learning testados, o modelo com melhor desempenho foi o Random Forest, que conseguiu classificar corretamente quase 80% da amostra de teste. O desempenho para classes com mais exemplos disponíveis no treino foi superior, e para a classe com menos instâncias apenas 71% dos registros que eram Good foram classificados como tal. Apesar de ser um desempenho satisfatório, o modelo não é ideal se para a área de negócios for muito importante identificar os clientes com o melhor score de créditos, para poder oferecer os serviços adequados a eles. Essa métrica pode ser melhorada futuramente com o uso de técnicas balanceamento de classes no treinamento, ou tentar encontrar hiperparâmetros mais adequados para esse cenário.

Ainda em relação ao modelo, o Dependence Plot para a variável de dívida restante na classe Poor mostrou que o modelo conseguiu aprender relações não lineares das variáveis, já que há um ponto em que o impacto de Outstanding_Debt cresce rapidamente na classe. Isso mostra a robustez do Random Forest em comparação a modelos de ML lineares.

Ao usar a técnica SHAP, verificou-se que as variáveis mais importantes para um cliente ser considerado com um score Poor são a dívida restante, a taxa de juros e os dias de atraso de pagamento da fatura. Valores altos para essas variáveis aumentam as chances da pessoa ter um score desse tipo, porém, para a dívida restante, após um certo valor a variável acaba não influenciando muito a decisão do modelo.

Para a classificação de Standard, o mix de créditos, a dívida restante e a taxa de juros foram as que mais influenciaram a decisão do modelo. Mix de créditos bons e taxas de juros altas reduzem a chance de classificação Standard, já dívidas restantes baixas aumentam a probabilidade de ser Standard.

Por fim, para a classificação Good, o mix de crédito bom e o pagamento de valor acima do mínimo foram contribuintes importantes para o modelo tomar essa decisão. Além disso, taxas de juros baixas e atrasos de pagamentos pequenos também ajudaram.

## Rcomendações práticas

Com os resultados proporcionados pela análise e uso de modelo de Machine Learning, é possível fazer algumas observações para instituições financeiras e seus clientes para que scores de crédito melhores sejam alcançados.

Para instituições financeiras, é importante que sejam priorizados os indicadores de comportamento financeiro dos clientes (juros, atrasos, dívidas) para a concessão de crédito, e como esses comportamentos podem mudar, é uma boa prática que os scores sejam atualizados periodicamente para evitar clientes que se tornem um risco.

Para os clientes dessas instituições, os resultados deste projeto indicam que as melhores formas de melhorar o score de créditos são: pagar as contas dentro do prazo, diminuir suas dívidas restantes, manter um bom mix de créditos e um histórico financeiro consistente ao longo do tempo.

## Como os resultados do modelo poderiam apoiar uma equipe de crédito na priorização de análises, concessão de crédito ou revisão de limites? 

Para apoiar as equipes de crédito, as classificações do modelo podem ajudar a direcionar um tipo de cliente para um tipo de avaliação mais adequada. Clientes mais prováveis de ser Good podem seguir para um fluxo de aprovação de crédito mais simplificado, enquanto clientes com maior probabilidade de serem Poor podem passar por uma análise de perfil mais criteriosa. Além disso, ele ajuda a automatizar a revisão periódica dos limites de créditos, visto que pode só ser executado novamente à medida que informações novas sobre um cliente são coletadas.

## Quais cuidados éticos e de negócio devem ser considerados antes de usar um modelo como esse em produção? 

Os dados dos clientes devem ser tratados de acordo com a LGPD, para que a instituição garanta o armazenamento seguro e uso das informações somente para finalidades autorizadas, reduzindo o risco de vazamento de dados sensíveis. Além disso, deve haver um cuidado para que o modelo não seja influenciado por vieses (que podem ser tirados dos dados de ocupação, por exemplo) e assim desfavorecer ou discriminar um perfil específico de cliente. Por fim, o modelo não deve substituir o critério humano na decisão de concessão de crédito, pois podem haver outros fatoes que influenciam no score de crédito dos clientes que não estão representados nos dados usados.
