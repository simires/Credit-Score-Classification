## 4. Tipos de dados do dataset de treino

A partir da execução de `df_train.info()` e do dicionário de dados, percebemos que alguns atributos estão com tipos de dados errados, o que pode indicar que há dados no formato incorreto ou valores nulos mascarados. As variáveis numéricas que aparecem como tipo string são:

- `Age` (discreta)
- `Annual_Income` (contínua)
- `Num_of_Loan` (discreta)
- `Num_of_Delayed_Payment` (discreta)
- `Changed_Credit_Limit` (contínua)
- `Outstanding_Debt` (contínua)
- `Amount_invested_monthly` (contínua)

Além disso, outras variáveis também estão com tipos incorretos:

- `Num_Credit_Inquiries` é variável numérica discreta (pois não há como fazer meia consulta de crédito), mas ela aparece como tipo float
- `Monthly_Balance` é variável numérica contínua, mas seu tipo aparece como object (sujeira nos dados)

Por fim, os valores das variáveis a seguir precisam passar por tratamento para contribuírem melhor para a análise:

- `Credit_History_Age` é um valor que aparece em formato de string, mas deve ser convertido posteriormente para um valor numérico (ex: quantidade de meses de histórico de crédito)
- `Payment_of_Min_Amount` é binária e está em formato textual. Ela pode ser convertida para 0 e 1

## 9. Distribuição da variável alvo 'Credit_Score'

```python
df_train["Credit_Score"].value_counts().to_frame("Quantidade")
```

```python
porcentagem = (
    df_train["Credit_Score"]
    .value_counts(normalize=True)
    * 100
)

plt.figure(figsize=(5,3))
ax = porcentagem.plot(kind="bar")

plt.title("Distribuição da variável alvo")
plt.ylabel("Porcentagem (%)")
plt.xticks(rotation=0)

for i, valor in enumerate(porcentagem):
    ax.text(i, valor/2, f"{valor:.1f}%", ha="center", color='white')

plt.show()
```

> A quantidade de tipos de score de crédito estão desbalanceadas, com a categoria Standard tendo 53,2% de registros, Poor com 29% e Good com 17,8%.
